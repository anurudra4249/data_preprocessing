```python
#deleting columns which have null values (greater than 20%)
missing_value_per = df.isnull().sum()/df.shape[0] * 100
missing_value_clm_gre_20 = missing_value_per[missing_value_per > 20].keys()
df2_drop_clm = df.drop(columns=missing_value_clm_gre_20)
df2_drop_clm.shape
```

## filling missing values by mean of each class for numeric data

```python
#classes in 'LotConfig' using unique() function
df['LotConfig'].unique()
```

```python
#we want to travel all rows and 'LotConfig'='Inside'
#replacing nan values by mean of particular class ('inside') in 'LotFrontage' column
df[df.loc[:,'LotConfig'] == "Inside"]["LotFrontage"].replace(np.nan,df[df.loc[:,'LotConfig'] == "Inside"]["LotFrontage"].mean())
```

```python
#using for loop for all classes
#coping dataframe
df_copy = df.copy()
for var_class in df['LotConfig'].unique():
    df_copy.update(df[df.loc[:,'LotConfig'] == var_class]["LotFrontage"].replace(np.nan,df[df.loc[:,'LotConfig'] == var_class]["LotFrontage"].mean()))
```

```python
df_copy.isnull().sum()
```

```python
#here we do all opration of numerical missing values which have null values lesser than 20%.
#zip() function use for zip data num_vars_miss and cat_vars
df_copy = df.copy()
num_vars_miss = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_vars = ['LotConfig','MasVnrType','GarageType']
for cat_var, num_var_miss in zip(cat_vars,num_vars_miss):
    for var_class in df[cat_var].unique():
        df_copy.update(df[df.loc[:,cat_var] == var_class][num_var_miss].replace(np.nan,df[df.loc[:,cat_var] == var_class][num_var_miss].mean()))
```

```python
df_copy[num_vars_miss].isnull().sum()
```

```python
#still some values not fill becoz of 'MasVnrType' have null values so 'MasVnrArea' still remain vacant. 
df_copy[df_copy[['MasVnrType']].isnull().any(axis=1)]
```

```python
#still some values not fill becoz of 'GarageType' have null values so 'GarageYrBlt' still remain vacant.
df_copy[df_copy[['GarageType']].isnull().any(axis=1)]
```

```python
#so, now we change categerical variables for 'MasVnrArea', 'GarageYrBlt'
df_copy = df.copy()
num_vars_miss = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_vars = ['LotConfig','Exterior2nd','KitchenQual']
for cat_var, num_var_miss in zip(cat_vars,num_vars_miss):
    for var_class in df[cat_var].unique():
        df_copy.update(df[df.loc[:,cat_var] == var_class][num_var_miss].replace(np.nan,df[df.loc[:,cat_var] == var_class][num_var_miss].mean()))
```

```python
df_copy[num_vars_miss].isnull().sum()
```

# Data Distribution

```python
# data visulisation before and after filling missing values.
plt.figure(figsize=(10,10))
for i, var in enumerate(num_vars_miss):
    plt.subplot(2,2,i+1)
    sns.distplot(df[var], bins=20,hist=False, kde_kws={'linewidth':8, 'color':'red'}, label="Original",)
    sns.distplot(df_copy[var], bins=20,hist=False, kde_kws={'linewidth':5, 'color':'green'},label="by_each_class_mean",)
    plt.legend()
```

# filling by Median of each class

```python
# filling values by median of each class
df_copy_median = df.copy()
num_vars_miss = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_vars = ['LotConfig','Exterior2nd','KitchenQual']
for cat_var, num_var_miss in zip(cat_vars,num_vars_miss):
    for var_class in df[cat_var].unique():
        df_copy_median.update(df[df.loc[:,cat_var] == var_class][num_var_miss].replace(np.nan,df[df.loc[:,cat_var] == var_class][num_var_miss].median()))
```

```python
df_copy_median[num_vars_miss].isnull().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(num_vars_miss):
    plt.subplot(2,2,i+1)
    sns.distplot(df[var], bins=20,hist=False, kde_kws={'linewidth':8, 'color':'red'}, label="Original")
    sns.distplot(df_copy[var], bins=20,hist=False, kde_kws={'linewidth':5, 'color':'green'},label="by_each_class_mean")
    sns.distplot(df_copy_median[var],hist=False, bins=20, kde_kws={'linewidth':3, 'color':'k'},label="by_each_class_median")
    plt.legend()
```
