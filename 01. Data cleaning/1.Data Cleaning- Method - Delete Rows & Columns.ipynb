## 1.first we deleting columns because most of null values present in particular columns.

```python
#data visualisation of null values columnswise
plt.figure(figsize=(25,25))
sns.heatmap(df.isnull())
```

```python
null_var = (df.isnull().sum()/df.shape[0]) *100
null_var
```

```python
#selecting columns which have null values greater than 17% of total values
drop_columns  = null_var[null_var >17].keys()
drop_columns
```

```python
#here we deleting columns which have null values greater than 17% of values
df2_drop_clm=df.drop(columns=drop_columns)
```

```python
df2_drop_clm.shape
```

```python
#data visulisation after deleting columns
plt.figure(figsize=(25,25))
sns.heatmap(df2_drop_clm.isnull())
```

## 2.  Delete rows which have null values. 

```python
#dropna function use for deleting row if any null value present in row
df3_drop_rows = df2_drop_clm.dropna()
```
