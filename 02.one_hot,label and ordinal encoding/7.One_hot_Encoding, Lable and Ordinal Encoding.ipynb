### Ordinal Encoding

```python
df2["KitchenQual"].value_counts()
```

```python
# """"""        KitchenQual: Kitchen qulality
#                  Ex---Excellent
#                  Gd---Good
#                  Ta---Typical/Average
#                  Fa---Fair
# """"""
order_label={"Ex":4,"Gd":3,"TA":2,"Fa":1}
```
